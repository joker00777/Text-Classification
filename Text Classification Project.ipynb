```python
def fill_dictionary(data,dictionary_words,stop_words):
    num_words=data.split()
    for word in num_words:
        if not word.lower() in stop_words:
            if(word.lower() in dictionary_words):
                #i am storing all the words in lower case so that they can be deleted 
                dictionary_words[word.lower()]=dictionary_words[word.lower()]+1     
            else:
                dictionary_words[word.lower()]=1

        
```

```python
def extract_words(x_train):
    dictionary_words={}
    #list of all stopwords
    stop_words=stopwords.words()
    for i in range(len(x_train)):
        #i am filling all words in dictionary 
        fill_dictionary(x_train[i][1],dictionary_words,stop_words)
    return dictionary_words    
```

```python
#this function is for indexing of features names in numpy array
def dict_for_features(dict_words):
    words_to_num={}
    i=0
    for word in dict_words:
        words_to_num[word]=i
        i=i+1
    return words_to_num    
```

```python
def printing_rows(x_row,dict_words,words_to_num):
    num_words=x_row.split()
    small_result=np.zeros(2000)
    for word in num_words:
        if word.lower() in dict_words:
            index=words_to_num[word.lower()]
            small_result[index]=small_result[index]+1
    return small_result    
```

```python
#ab mere paas meri final dictionary aa chuki hain
def data_frame(x_train,dict_words,words_to_num):
    index=0
    result=np.zeros((len(x_train),2000))
    for i in range(len(x_train)):
        #small_result=np.array([])
        small_result=printing_rows(x_train[i][1],dict_words,words_to_num)
        result[index]=small_result
        index=index+1
    return result  
```

```python
#fit function which is creating dictionary of dictionaries
def fit(x_train_result,y_train):
    result={}
    class_values=set(y_train)
    for current_class in class_values:
        result[current_class]={}
        result["total_data"]=x_train_result[:,:].sum()
        current_class_rows=(y_train==current_class)
        x_train_current=x_train_result[current_class_rows]
        y_train_current=y_train[current_class_rows]
        num_features=x_train_result.shape[1]
        result[current_class]["total_count"]=x_train_current[:,:].sum()
        for j in range(0,num_features):
            result[current_class][j]=x_train_current[:,j].sum()
    return result        
            
```

```python
#this will return me the probability of y1 for that input
def probability(x,dict_fit,current_class):
        log_prob = np.log(dict_fit[current_class]['total_count']) - np.log(dict_fit['total_data'])
        total_words = len(x)
        for i in range(len(x)):
            current_word_prob = x[i]*(np.log(dict_fit[current_class][i]+1)-np.log(dict_fit[current_class]['total_count']+total_words))
            log_prob += current_word_prob
        return log_prob
```

```python
#this will give me the best probablity for single input of x
def predictsinglepoint(x,dict_fit):
    classes=dict_fit.keys()
    best_p=-1000
    best_class=''
    first_run=True
    for current_class in classes:
        if current_class=="total_data":
            continue
        p_current_class=probability(x,dict_fit,current_class)
        if first_run==True or p_current_class>best_p:
            best_p=p_current_class
            best_class=current_class
        first_run=False    
    return best_class       
        
        
    
```

```python
#predict function will predict the output for us
def predict(x_test_data,dict_fit):
    y_predict=[]
    for x in x_test_data:
        y_class=predictsinglepoint(x,dict_fit)
        y_predict.append(y_class)
    return y_predict    
```
